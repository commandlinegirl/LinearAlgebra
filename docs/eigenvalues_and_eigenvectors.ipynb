```python
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler
from matplotlib import rcParams

rcParams['text.usetex'] = True
rcParams['figure.figsize'] = 15, 5

def set_ax(ax, vectors):
  ax.grid()
  ax.axis('equal')
  ax.arrow(0, 0, vectors[0,0], vectors[1,0], head_width=0.05, length_includes_head=True, head_length=0.1, fc='r', ec='r')
  ax.arrow(0, 0, vectors[0,1], vectors[1,1], head_width=0.05, length_includes_head=True, head_length=0.1, fc='k', ec='k')
  ax.set_xlim([-8, 8])
  ax.set_ylim([-8, 8])

def plot_example(A, annotation):
  # the column eigvectors[:,i] is the eigenvector corresponding to the eigvalues[i].
  eigvalues, eigvectors = np.linalg.eig(A)
  print("eigvals:\n[{}]".format(eigvalues))
  print("eigvectors:\n[{}]".format(eigvectors))
  t = np.linspace(0, np.pi*2, 100)
  v = np.array([np.cos(t), np.sin(t)])
  w = np.dot(A, v)
  scaled = A.T * eigvectors

  f, (ax1, ax2) = plt.subplots(1, 2, sharey=False)
  # Original space
  ax1.set_title('Original vector space with matrix eigenvectors')
  set_ax(ax1, eigvectors)
  ax1.annotate(annotation,
               (-7.5, 2.5), size=20)
  ax1.plot(v[0], v[1], linewidth=1)

  # Space multiplied by the matrix
  ax2.set_title('Vector space transformed by the matrix')
  set_ax(ax2, scaled)
  ax2.plot(w[0].T, w[1].T, linewidth=1)

  plt.show()

#####################################
print("Example 1")
plot_example(np.matrix('2 3; 3 1'),
             r"$ matrix: \left[ \begin{array}{ccc} 2 & 3 \\ 3 & 1  \end{array} \right] $")

print("Example 2")
plot_example(np.matrix('-2 3; 3 -2'),
             r"$ matrix: \left[ \begin{array}{ccc} -2 & 3 \\ 3 & -2  \end{array} \right] $")

print("Example 3")
plot_example(np.matrix('1 3; 3 1'),
             r"$ matrix: \left[ \begin{array}{ccc} 1 & 3 \\ 3 & 1  \end{array} \right] $")

print("Example 4")
plot_example(np.matrix('4 1; 1 4'),
             r"$ matrix: \left[ \begin{array}{ccc} 4 & 1 \\ 1 & 4  \end{array} \right] $")
```
